```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 设置 Matplotlib 和 Seaborn 的样式，让图表更美观
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用于显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题
sns.set_style("whitegrid") # 设置Seaborn风格

# --- 1. 数据加载：从GitHub直接导入 ---
github_raw_url = 'https://raw.githubusercontent.com/Astraeushub/Astraeushub/main/%E9%99%84%E4%BB%B61.xlsx'

try:
    df = pd.read_excel(github_raw_url)
    print("数据加载成功！")

except Exception as e:
    print(f"错误：从GitHub加载文件失败。请检查URL或网络连接。\n错误信息：{e}")
    exit() # 如果加载失败，停止执行

# --- 2. 数据清洗与初步处理 ---

# 2.1 删除多余的 'Unnamed' 列
# 找到所有以 'Unnamed:' 开头的列
unnamed_cols = [col for col in df.columns if 'Unnamed:' in col]
df.drop(columns=unnamed_cols, inplace=True)
print("\n已删除 'Unnamed' 列。")

# 2.2 清洗列名：去除括号和中文描述，简化名称
df.rename(columns={
    '妊娠时间（周数）': '妊娠时间',
    '整晚睡眠时间（时：分：秒）': '整晚睡眠时间',
    '婴儿年龄（月）': '婴儿年龄'
}, inplace=True)
# 顺便检查并去除所有列名中的空格
df.columns = df.columns.str.strip()
print("\n已简化列名：")
print(df.columns.tolist())


# 2.3 处理 '整晚睡眠时间' 列：从字符串 'HH:MM:SS' 转换为浮点小时数
def convert_time_to_hours(time_str):
    if pd.isna(time_str): # 处理NaN值
        return np.nan
    try:
        # 假设时间格式始终是 HH:MM:SS
        parts = str(time_str).split(':')
        hours = int(parts[0])
        minutes = int(parts[1])
        # seconds = int(parts[2]) # 秒数通常可以忽略
        return hours + minutes / 60
    except:
        return np.nan # 转换失败也返回NaN

df['整晚睡眠时间'] = df['整晚睡眠时间'].apply(convert_time_to_hours)
print("\n'整晚睡眠时间'已转换为小时数。")


# 2.4 处理缺失值 (Problem 1 仅使用完整数据)
# 题目要求最后20组婴儿行为信息缺失，且我们看到睡眠相关数据也缺失
# 问题1是研究规律，通常需要完整数据。
# 因此，我们先删除所有关键分析列的缺失值。
# 关键分析列包括：母亲心理/身体指标, 婴儿行为, 婴儿睡眠指标
critical_cols_for_analysis = [
    '母亲年龄', '婚姻状况', '教育程度', '妊娠时间', '分娩方式',
    'CBTS', 'EPDS', 'HADS', '婴儿行为特征', '婴儿性别', '婴儿年龄',
    '整晚睡眠时间', '睡醒次数', '入睡方式'
]

initial_rows = df.shape[0]
# 仅保留关键列中非缺失的行
df_cleaned = df.dropna(subset=critical_cols_for_analysis).copy() # 使用.copy()避免SettingWithCopyWarning
rows_after_dropna = df_cleaned.shape[0]

print(f"\n原始数据行数：{initial_rows}")
print(f"删除关键列缺失值后，用于问题1分析的数据行数：{rows_after_dropna}")
print(f"删除了 {initial_rows - rows_after_dropna} 行（主要为391-400号婴儿的数据以及可能的数据尾部杂质）。")

# 确认用于分析的数据范围 (应该从1到380号)
print("\n用于问题1分析的数据前5行：")
print(df_cleaned.head())
print("\n用于问题1分析的数据后5行：")
print(df_cleaned.tail())
print("\n用于问题1分析的数据列信息：")
df_cleaned.info()

# 2.5 检查 '婴儿行为特征' 列的唯一值，确保是预期的
print("\n'婴儿行为特征'列的唯一值 (清洗后):")
print(df_cleaned['婴儿行为特征'].unique())
# 检查 '入睡方式' 的唯一值，看看是否有1-5以外的数字
print("\n'入睡方式'列的唯一值 (清洗后):", df_cleaned['入睡方式'].unique())


# --- 重要：将处理后的DataFrame用于后续分析 ---
# 从现在开始，我们所有的分析都将基于 df_cleaned
# （或者如果您需要，可以保持原始df以便于问题2处理缺失值）
# 建议为每个问题维护一个合适的DataFrame版本。
# 这里我们用 df_cleaned 作为问题1的输入。
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 确保 Matplotlib 和 Seaborn 的样式设置已经到位 (如果之前已经运行过，可以跳过这部分，但重复设置也无妨)
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用于显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题
sns.set_style("whitegrid") # 设置Seaborn风格

# --- 请确保 df_processed_onehot 是您最新处理好的DataFrame ---
# 假设df_processed_onehot已经包含了所有独热编码和清洗后的数据
# 如果您在Jupyter Notebook中是连续运行，这个变量应该已经存在。
# 如果是重新启动Notebook，请确保重新运行所有前面的数据加载和处理步骤。

print("\n--- 步骤 5：数值型变量相关性分析 ---")

# 定义母亲的数值型指标
mother_numerical_features = ['母亲年龄', '妊娠时间', 'CBTS', 'EPDS', 'HADS']

# 定义婴儿的数值型睡眠指标
baby_sleep_numerical_targets = ['整晚睡眠时间', '睡醒次数']

# 合并所有相关数值列，计算它们之间的相关系数矩阵
# 确保所有这些列都存在于 df_processed_onehot 中
# 剔除'编号'列，因为它不是一个用于相关性分析的数值特征
all_numerical_cols = [col for col in mother_numerical_features + baby_sleep_numerical_targets if col in df_processed_onehot.columns]

# 确保所有列都是数值类型，以防万一
for col in all_numerical_cols:
    df_processed_onehot[col] = pd.to_numeric(df_processed_onehot[col], errors='coerce')

# 再次处理可能的NaN（尽管我们在前面对df_cleaned做了dropna，但类型转换可能产生）
# 仅针对我们要计算相关性的这些数值列，再次删除NaN，确保corr()函数不受影响
df_analysis_numerical = df_processed_onehot[all_numerical_cols].dropna()
if df_analysis_numerical.shape[0] < df_processed_onehot.shape[0]:
    print(f"警告：为相关性分析删除了 {df_processed_onehot.shape[0] - df_analysis_numerical.shape[0]} 行缺失数据。")


correlation_matrix = df_analysis_numerical.corr()

print("\n数值型变量相关系数矩阵：")
print(correlation_matrix)

# 使用热力图可视化相关系数矩阵
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
plt.title('母亲及婴儿数值型指标相关性热力图')
plt.show()

# 进一步分析与婴儿睡眠相关的具体相关性
print("\n婴儿睡眠指标与母亲数值型指标的相关性：")
for target_col in baby_sleep_numerical_targets:
    if target_col in correlation_matrix.columns:
        print(f"\n--- {target_col} ---")
        # 筛选出与当前睡眠指标相关性较高的母亲指标
        # 按照相关系数的绝对值进行降序排序，以突出最强的关系
        relevant_correlations = correlation_matrix[target_col][mother_numerical_features].sort_values(ascending=False, key=abs)
        print(relevant_correlations)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 确保 Matplotlib 和 Seaborn 的样式设置已经到位
plt.rcParams['font.sans-serif'] = ['SimHei'] # 用于显示中文标签
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题
sns.set_style("whitegrid") # 设置Seaborn风格

# --- 确保 df_cleaned 和 df_processed_onehot 是您最新处理好的DataFrame ---
# df_cleaned 包含原始数字编码的分类列，用于分组和绘图标签
# df_processed_onehot 包含独热编码后的所有列和数值目标列

print("\n--- 步骤 6：分类变量与数值型变量关系分析 ---")

# 母亲的分类指标 (使用原始的数字编码列来作为分组依据，因为BoxPlot需要类别标签)
mother_categorical_cols_original = [
    '婚姻状况', '教育程度', '分娩方式', '婴儿性别', '入睡方式'
]

# 婴儿的数值型睡眠指标 和 母亲的心理指标
target_numerical_cols = [
    '整晚睡眠时间', '睡醒次数', 'CBTS', 'EPDS', 'HADS'
]

print("\n分类变量对数值型变量的影响：")

# 辅助映射字典 (与之前定义的保持一致)
edu_map = {1: '小学', 2: '初中', 3: '高中', 4: '大学', 5: '研究生'}
marital_map = {1: '未婚', 2: '已婚'}
delivery_map = {1: '自然分娩', 2: '剖宫产'}
gender_map = {1: '男性', 2: '女性'}
sleep_mode_map = {1: '哄睡法', 2: '抚触法', 3: '安抚奶嘴法', 4: '环境营造法', 5: '定时法'}

# 将映射字典整合到一个更大的字典中，方便查找
all_maps = {
    '婚姻状况': marital_map,
    '教育程度': edu_map,
    '分娩方式': delivery_map,
    '婴儿性别': gender_map,
    '入睡方式': sleep_mode_map
}

for cat_col_orig in mother_categorical_cols_original:
    if cat_col_orig not in df_cleaned.columns: # 确保原始列存在于df_cleaned中
        print(f"警告：原始数据中找不到列 '{cat_col_orig}'，跳过。")
        continue

    # 获取当前分类列的映射字典
    current_map = all_maps.get(cat_col_orig, {}) # 如果没有映射，就用空字典

    for target_num_col in target_numerical_cols:
        # 使用 df_cleaned 的原始分类列和 df_processed_onehot 的数值目标列
        # 注意：这里确保用于绘图的x轴数据是分类类型，并且y轴数据是数值类型
        # Pandas的boxplot可以智能处理
        
        # 仅当目标列存在时才绘图
        if target_num_col not in df_processed_onehot.columns:
            continue

        plt.figure(figsize=(8, 6))
        
        # 使用原始 df_cleaned 中的分类列来分组，并应用映射使标签可读
        # 如果原始df_cleaned的列名与all_maps的键不完全匹配，可能需要调整
        # 确保 df_cleaned[cat_col_orig] 是原始数字编码的列
        x_values_for_plot = df_cleaned[cat_col_orig].map(current_map).fillna(df_cleaned[cat_col_orig]).astype(str) # 转换为字符串类型，确保seaborn按分类处理

        sns.boxplot(x=x_values_for_plot, y=df_processed_onehot[target_num_col])
        plt.title(f'{target_num_col} 与 {cat_col_orig} 的关系')
        plt.xlabel(cat_col_orig)
        plt.ylabel(target_num_col)
        plt.xticks(rotation=45, ha='right') # 旋转X轴标签以避免重叠
        plt.tight_layout() # 自动调整布局，避免标签截断
        plt.show()

        # 同时输出分组统计均值
        print(f"\n--- '{cat_col_orig}' 对 '{target_num_col}' 的影响 (平均值) ---")
        grouped_means = df_cleaned.groupby(cat_col_orig)[target_num_col].mean()
        # 尝试应用映射，使输出更可读
        if current_map:
            grouped_means.index = grouped_means.index.map(current_map)
        print(grouped_means)
        print("-" * 60) # 分隔线
```
