```python
import pandas as pd
import numpy as np
from scipy.optimize import nnls
import matplotlib.pyplot as plt
import matplotlib as mpl

# --- 修复中文字符乱码 ---
mpl.rcParams['font.sans-serif'] = # 修正：添加了字体名称
mpl.rcParams['axes.unicode_minus'] = False

# --- 文件路径和工作表名称 ---
github_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
sun_spd_sheet = 'Problem 3 SUN_SPD'
led_spd_sheet = 'Problem 2_LED_SPD'

# --- 1. 数据读取与修正后的预处理 ---
try:
    # 修正：设置波长列为索引，并清除列名中的括号
    df_sun_spd = pd.read_excel(github_url, sheet_name=sun_spd_sheet, index_col=0)
    df_sun_spd.index = df_sun_spd.index.str.extract(r'(\d+)', expand=False).astype(int)
    
    df_led_spd = pd.read_excel(github_url, sheet_name=led_spd_sheet, index_col=0)
    df_led_spd.index = df_led_spd.index.str.extract(r'(\d+)', expand=False).astype(int)

    print("数据读取与预处理成功。")
except Exception as e:
    print(f"读取文件或预处理失败，请检查链接或数据格式: {e}")
    exit()

# 确保两个数据框波长范围一致
min_wl = max(df_sun_spd.index.min(), df_led_spd.index.min())
max_wl = min(df_sun_spd.index.max(), df_led_spd.index.max())
common_wavelengths = np.arange(min_wl, max_wl + 1)

df_sun_spd = df_sun_spd.reindex(common_wavelengths).interpolate().dropna(how='all', axis=0)
df_led_spd = df_led_spd.reindex(common_wavelengths).interpolate().dropna(how='all', axis=0)

# --- 2.3. 建立并求解优化模型 ---
# 将LED光谱数据转换为矩阵
A = df_led_spd.values

# 存储所有时间点的LED光强控制参数
control_strategy = pd.DataFrame(index=df_sun_spd.columns, columns=df_led_spd.columns)

for time_point in df_sun_spd.columns:
    b = df_sun_spd[time_point].values
    
    # 解决非负最小二乘问题，找到最佳光强系数
    x, _ = nnls(A, b)
    control_strategy.loc[time_point] = x

print("\n--- 全天LED光强控制策略（部分展示）：---")
print(control_strategy.head())

# --- 模型评估代码开始 ---
print("\n--- 模型评估开始 ---")

# 存储每个时间点的RMSE和R-squared
rmse_values =
r_squared_values =
time_points_list = df_sun_spd.columns.tolist()

for time_point in time_points_list:
    target_spd = df_sun_spd[time_point].values
    led_intensities = control_strategy.loc[time_point].values
    synthesized_spd = A @ led_intensities
    
    # 计算RMSE
    rmse = np.sqrt(np.mean((target_spd - synthesized_spd)**2))
    rmse_values.append(rmse)
    
    # 计算R-squared
    ss_total = np.sum((target_spd - np.mean(target_spd))**2)
    ss_residual = np.sum((target_spd - synthesized_spd)**2)
    if ss_total > 0: # 避免除以零
        r_squared = 1 - (ss_residual / ss_total)
    else:
        r_squared = 1.0 # 如果目标SPD是常数，且模型完美拟合
    r_squared_values.append(r_squared)

# 创建评估结果DataFrame
evaluation_df = pd.DataFrame({
    '时间点': time_points_list,
    'RMSE': rmse_values,
    'R-squared': r_squared_values
})

print("\n--- 全天候光谱拟合评估结果：---")
print(evaluation_df)

print(f"\n平均RMSE: {np.mean(rmse_values):.4f}")
print(f"平均R-squared: {np.mean(r_squared_values):.4f}")

# 绘制RMSE随时间变化图 (对应论文中的图1)
plt.figure(figsize=(12, 6))
plt.plot(time_points_list, rmse_values, marker='o', linestyle='-', color='red')
plt.title('全天候光谱拟合均方根误差（RMSE）随时间变化图', fontsize=16)
plt.xlabel('时间点')
plt.ylabel('RMSE')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('rmse_over_time.png')


# 绘制R-squared随时间变化图
plt.figure(figsize=(12, 6))
plt.plot(time_points_list, r_squared_values, marker='o', linestyle='-', color='green')
plt.title('全天候光谱拟合R-squared随时间变化图', fontsize=16)
plt.xlabel('时间点')
plt.ylabel('R-squared')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.savefig('r_squared_over_time.png')

print("\n--- 模型评估结束 ---")
# --- 模型评估代码结束 ---
```

```python
# 导入必要的库
import pandas as pd
import numpy as np
from scipy.optimize import nnls
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 设置matplotlib字体以支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

# 定义一个函数，用于计算均方根误差 (RMSE)
def calculate_rmse(target, predicted):
    """
    计算两个光谱之间的均方根误差。

    参数:
    target (np.array): 目标光谱数据。
    predicted (np.array): 预测光谱数据。

    返回:
    float: 均方根误差。
    """
    return np.sqrt(np.mean((target - predicted)**2))

def run_problem3_analysis():
    """
    执行问题3的模型评估和可视化。
    该函数将模型匹配全天太阳光谱的鲁棒性进行分析。
    """
    print("--- 正在进行问题3：模型鲁棒性评估与可视化 ---")
    
    # 1. 读取数据
    try:
        df_led_spd = pd.read_csv('附录.xlsx - Problem 2_LED_SPD.csv')
        df_sun_spd = pd.read_csv('附录.xlsx - Problem 3 SUN_SPD.csv')
    except FileNotFoundError:
        print("错误：未找到问题3所需的数据文件。请确保 '附录.xlsx - Problem 2_LED_SPD.csv' 和 '附录.xlsx - Problem 3 SUN_SPD.csv' 存在于当前目录。")
        return

    # 数据预处理
    wavelengths = df_led_spd.iloc[:, 0]
    led_spds = df_led_spd.iloc[:, 1:].to_numpy()
    sun_times = df_sun_spd.columns[1:]
    sun_spds = df_sun_spd.iloc[:, 1:].to_numpy()

    # 确保波长一致性
    if not np.allclose(wavelengths, df_sun_spd.iloc[:, 0]):
        print("警告：LED和太阳光谱的波长不一致，正在进行插值处理...")
        from scipy.interpolate import interp1d
        sun_spds_interp = []
        for i in range(sun_spds.shape[1]):
            f = interp1d(df_sun_spd.iloc[:, 0], sun_spds[:, i])
            sun_spds_interp.append(f(wavelengths))
        sun_spds = np.array(sun_spds_interp).T

    # 2. NNLS光谱优化模型
    A = led_spds
    rmse_values = []
    optimal_x_values = []

    for i in range(sun_spds.shape[1]):
        b = sun_spds[:, i]
        # 使用NNLS求解最优混合比例
        x, _ = nnls(A, b)
        optimal_x_values.append(x)
        # 合成光谱
        synthesized_spd = np.dot(A, x)
        # 计算RMSE
        rmse = calculate_rmse(b, synthesized_spd)
        rmse_values.append(rmse)

    # 3. 结果可视化：RMSE随时间变化的折线图
    plt.figure(figsize=(10, 6))
    plt.plot(sun_times, rmse_values, marker='o')
    plt.title('模型匹配鲁棒性评估：RMSE随时间的变化')
    plt.xlabel('时间')
    plt.ylabel('均方根误差 (RMSE)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("模型鲁棒性评估图表已生成。")

    # 4. 结果可视化：某一时刻的太阳光谱与合成光谱对比
    # 选择一个时间点进行展示（例如中午）
    midday_index = sun_times.get_loc('12:30:00')
    target_spd = sun_spds[:, midday_index]
    synthesized_spd = np.dot(A, optimal_x_values[midday_index])
    
    plt.figure(figsize=(10, 6))
    plt.plot(wavelengths, target_spd, label=f'目标太阳光谱 ({sun_times[midday_index]})')
    plt.plot(wavelengths, synthesized_spd, '--', label=f'模型合成光谱', linewidth=2)
    plt.title('光谱匹配效果：太阳光谱与合成光谱对比')
    plt.xlabel('波长 (nm)')
    plt.ylabel('相对光强')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("光谱匹配效果对比图表已生成。")
    print("-" * 50)

def run_problem4_analysis():
    """
    执行问题4的睡眠指标分析与可视化。
    该函数将计算各项睡眠指标，并进行统计检验和箱线图绘制。
    """
    print("--- 正在进行问题4：睡眠指标分析与可视化 ---")
    
    # 1. 读取数据
    try:
        df_raw = pd.read_csv('附录.xlsx - Problem 4.csv')
    except FileNotFoundError:
        print("错误：未找到问题4所需的数据文件。请确保 '附录.xlsx - Problem 4.csv' 存在于当前目录。")
        return
    
    # 2. 数据预处理 - 转换为长格式，方便分析
    # 假设每行代表一个30秒的睡眠分期数据
    num_subjects = 11
    nights_per_subject = 3
    
    data_list = []
    # 遍历每个被试和每个夜晚
    for subject_num in range(1, num_subjects + 1):
        for night_num in range(1, nights_per_subject + 1):
            col_name = f'被试{subject_num}.{night_num}'
            # 找到对应列
            # 原始文件头 '被试1' 后面是三个夜，所以每被试3列
            start_col_index = (subject_num - 1) * nights_per_subject + night_num
            col_data = df_raw.iloc[:, start_col_index]

            if pd.notnull(col_data).any():
                # 创建临时DataFrame，用于存储当前被试-夜晚的数据
                temp_df = pd.DataFrame({
                    'subject': f'被试{subject_num}',
                    'night': f'Night {night_num}',
                    'sleep_stage': col_data
                })
                data_list.append(temp_df)

    df_long = pd.concat(data_list, ignore_index=True)
    df_long = df_long.dropna()
    df_long['sleep_stage'] = df_long['sleep_stage'].astype(int)

    # 3. 计算各项睡眠指标
    # 假设每行数据代表一个30秒的睡眠周期
    epoch_duration_seconds = 30
    
    metrics = []
    # 假设光照条件与夜晚的映射关系
    # Night 1: 普通光照
    # Night 2: 黑暗环境
    # Night 3: 优化光照
    # 这是根据用户之前提供的分析结果（优化光照效果最好）做出的一个合理假设。
    condition_mapping = {
        'Night 1': '普通光照',
        'Night 2': '黑暗环境',
        'Night 3': '优化光照'
    }

    for subject, group_df in df_long.groupby('subject'):
        for night, night_df in group_df.groupby('night'):
            # 总睡眠时间 (Total Sleep Time, TST)
            # 睡眠分期4代表清醒，其他都是睡眠
            is_sleep = night_df['sleep_stage'] != 4
            tst = is_sleep.sum() * epoch_duration_seconds / 3600  # 转换为小时
            
            # 睡眠效率 (Sleep Efficiency, SE)
            total_epochs = len(night_df)
            se = (is_sleep.sum() / total_epochs) * 100 if total_epochs > 0 else 0
            
            # 入睡潜伏期 (Sleep Latency, SOL)
            # 从数据开始到第一个非清醒状态（sleep_stage != 4）的第一个周期
            first_sleep_epoch = night_df[is_sleep].index.min()
            start_index = night_df.index.min()
            sol = (first_sleep_epoch - start_index) * epoch_duration_seconds / 60 if pd.notnull(first_sleep_epoch) else np.nan # 转换为分钟
            
            metrics.append({
                '被试': subject,
                '夜晚': night,
                '光照条件': condition_mapping.get(night),
                '总睡眠时间 (小时)': tst,
                '睡眠效率 (%)': se,
                '入睡潜伏期 (分钟)': sol
            })

    df_metrics = pd.DataFrame(metrics)
    
    # 4. 结果可视化：箱线图
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle('不同光照条件下睡眠指标的分布', fontsize=16)
    
    # 设置x轴顺序，以匹配之前Tukey HSD的输出
    order = ['普通光照', '黑暗环境', '优化光照']

    for i, metric in enumerate(['总睡眠时间 (小时)', '睡眠效率 (%)', '入睡潜伏期 (分钟)']):
        # 创建箱线图
        df_metrics.boxplot(column=metric, by='光照条件', ax=axes[i], order=order)
        axes[i].set_title(metric)
        axes[i].set_xlabel('光照条件')
        axes[i].set_ylabel(metric)
        axes[i].get_figure().suptitle('')
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    print("睡眠指标箱线图已生成。")

    # 5. 统计检验（ANOVA & Tukey's HSD）
    print("\n--- 正在进行统计检验 ---")
    
    for metric in ['总睡眠时间 (小时)', '睡眠效率 (%)', '入睡潜伏期 (分钟)']:
        print(f"\n--- 对 '{metric}' 进行方差分析 (ANOVA) ---")
        
        group1 = df_metrics[df_metrics['光照条件'] == '普通光照'][metric].dropna()
        group2 = df_metrics[df_metrics['光照条件'] == '黑暗环境'][metric].dropna()
        group3 = df_metrics[df_metrics['光照条件'] == '优化光照'][metric].dropna()
        
        # 移除空值以避免错误
        if len(group1) > 1 and len(group2) > 1 and len(group3) > 1:
            f_statistic, p_value = stats.f_oneway(group1, group2, group3)
            print(f"F-statistic: {f_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")
            
            if p_value < 0.05:
                print("结果解读：P-value 小于显著性水平 0.05，结论：三种光照环境对该指标的影响存在显著差异。")
                print("\n--- 进行事后检验 (Tukey's HSD) ---")
                
                # 事后检验需要扁平化的数据
                data_for_tukey = df_metrics[metric].dropna()
                groups_for_tukey = df_metrics['光照条件'].loc[data_for_tukey.index]
                
                tukey_result = pairwise_tukeyhsd(data_for_tukey, groups_for_tukey, alpha=0.05)
                print(tukey_result)
            else:
                print("结果解读：P-value 大于显著性水平 0.05，结论：三种光照环境对该指标的影响不存在显著差异。")
        else:
            print(f"数据不足，无法对 {metric} 进行统计分析。")

    print("-" * 50)

# 主程序入口
if __name__ == '__main__':
    run_problem3_analysis()
    run_problem4_analysis()

```

```python
# 导入必要的库
import pandas as pd
import numpy as np
from scipy.optimize import nnls
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

# 设置matplotlib字体以支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题

# 定义一个函数，用于计算均方根误差 (RMSE)
def calculate_rmse(target, predicted):
    """
    计算两个光谱之间的均方根误差。

    参数:
    target (np.array): 目标光谱数据。
    predicted (np.array): 预测光谱数据。

    返回:
    float: 均方根误差。
    """
    return np.sqrt(np.mean((target - predicted)**2))

def run_problem3_analysis():
    """
    执行问题3的模型评估和可视化。
    该函数将模型匹配全天太阳光谱的鲁棒性进行分析。
    """
    print("--- 正在进行问题3：模型鲁棒性评估与可视化 ---")
    
    # 1. 读取数据，使用您提供的URL
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_led_spd = pd.read_excel(file_url, sheet_name='Problem 2_LED_SPD')
        df_sun_spd = pd.read_excel(file_url, sheet_name='Problem 3 SUN_SPD')
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 数据预处理
    wavelengths = df_led_spd.iloc[:, 0]
    led_spds = df_led_spd.iloc[:, 1:].to_numpy()
    sun_times = df_sun_spd.columns[1:]
    sun_spds = df_sun_spd.iloc[:, 1:].to_numpy()

    # 确保波长一致性
    if not np.allclose(wavelengths, df_sun_spd.iloc[:, 0]):
        print("警告：LED和太阳光谱的波长不一致，正在进行插值处理...")
        from scipy.interpolate import interp1d
        sun_spds_interp = []
        for i in range(sun_spds.shape[1]):
            f = interp1d(df_sun_spd.iloc[:, 0], sun_spds[:, i])
            sun_spds_interp.append(f(wavelengths))
        sun_spds = np.array(sun_spds_interp).T

    # 2. NNLS光谱优化模型
    A = led_spds
    rmse_values = []
    optimal_x_values = []

    for i in range(sun_spds.shape[1]):
        b = sun_spds[:, i]
        # 使用NNLS求解最优混合比例
        x, _ = nnls(A, b)
        optimal_x_values.append(x)
        # 合成光谱
        synthesized_spd = np.dot(A, x)
        # 计算RMSE
        rmse = calculate_rmse(b, synthesized_spd)
        rmse_values.append(rmse)

    # 3. 结果可视化：RMSE随时间变化的折线图
    plt.figure(figsize=(10, 6))
    plt.plot(sun_times, rmse_values, marker='o')
    plt.title('模型匹配鲁棒性评估：RMSE随时间的变化')
    plt.xlabel('时间')
    plt.ylabel('均方根误差 (RMSE)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("模型鲁棒性评估图表已生成。")

    # 4. 结果可视化：某一时刻的太阳光谱与合成光谱对比
    # 选择一个时间点进行展示（例如中午）
    midday_index = sun_times.get_loc('12:30:00')
    target_spd = sun_spds[:, midday_index]
    synthesized_spd = np.dot(A, optimal_x_values[midday_index])
    
    plt.figure(figsize=(10, 6))
    plt.plot(wavelengths, target_spd, label=f'目标太阳光谱 ({sun_times[midday_index]})')
    plt.plot(wavelengths, synthesized_spd, '--', label=f'模型合成光谱', linewidth=2)
    plt.title('光谱匹配效果：太阳光谱与合成光谱对比')
    plt.xlabel('波长 (nm)')
    plt.ylabel('相对光强')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("光谱匹配效果对比图表已生成。")
    print("-" * 50)

def run_problem4_analysis():
    """
    执行问题4的睡眠指标分析与可视化。
    该函数将计算各项睡眠指标，并进行统计检验和箱线图绘制。
    """
    print("--- 正在进行问题4：睡眠指标分析与可视化 ---")
    
    # 1. 读取数据，使用您提供的URL
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_raw = pd.read_excel(file_url, sheet_name='Problem 4')
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return
    
    # 2. 数据预处理 - 转换为长格式，方便分析
    # 假设每行代表一个30秒的睡眠分期数据
    num_subjects = 11
    nights_per_subject = 3
    
    data_list = []
    # 遍历每个被试和每个夜晚
    for subject_num in range(1, num_subjects + 1):
        for night_num in range(1, nights_per_subject + 1):
            # 原始文件头 '被试1' 后面是三个夜，所以每被试3列
            start_col_index = (subject_num - 1) * nights_per_subject + night_num
            col_data = df_raw.iloc[:, start_col_index]

            if pd.notnull(col_data).any():
                # 创建临时DataFrame，用于存储当前被试-夜晚的数据
                temp_df = pd.DataFrame({
                    'subject': f'被试{subject_num}',
                    'night': f'Night {night_num}',
                    'sleep_stage': col_data
                })
                data_list.append(temp_df)

    df_long = pd.concat(data_list, ignore_index=True)
    df_long = df_long.dropna()
    df_long['sleep_stage'] = df_long['sleep_stage'].astype(int)

    # 3. 计算各项睡眠指标
    # 假设每行数据代表一个30秒的睡眠周期
    epoch_duration_seconds = 30
    
    metrics = []
    # 假设光照条件与夜晚的映射关系
    # Night 1: 普通光照
    # Night 2: 黑暗环境
    # Night 3: 优化光照
    # 这是根据用户之前提供的分析结果（优化光照效果最好）做出的一个合理假设。
    condition_mapping = {
        'Night 1': '普通光照',
        'Night 2': '黑暗环境',
        'Night 3': '优化光照'
    }

    for subject, group_df in df_long.groupby('subject'):
        for night, night_df in group_df.groupby('night'):
            # 总睡眠时间 (Total Sleep Time, TST)
            # 睡眠分期4代表清醒，其他都是睡眠
            is_sleep = night_df['sleep_stage'] != 4
            tst = is_sleep.sum() * epoch_duration_seconds / 3600  # 转换为小时
            
            # 睡眠效率 (Sleep Efficiency, SE)
            total_epochs = len(night_df)
            se = (is_sleep.sum() / total_epochs) * 100 if total_epochs > 0 else 0
            
            # 入睡潜伏期 (Sleep Latency, SOL)
            # 从数据开始到第一个非清醒状态（sleep_stage != 4）的第一个周期
            first_sleep_epoch = night_df[is_sleep].index.min()
            start_index = night_df.index.min()
            sol = (first_sleep_epoch - start_index) * epoch_duration_seconds / 60 if pd.notnull(first_sleep_epoch) else np.nan # 转换为分钟
            
            metrics.append({
                '被试': subject,
                '夜晚': night,
                '光照条件': condition_mapping.get(night),
                '总睡眠时间 (小时)': tst,
                '睡眠效率 (%)': se,
                '入睡潜伏期 (分钟)': sol
            })

    df_metrics = pd.DataFrame(metrics)
    
    # 4. 结果可视化：箱线图
    fig, axes = plt.subplots(1, 3, figsize=(18, 6))
    fig.suptitle('不同光照条件下睡眠指标的分布', fontsize=16)
    
    # 设置x轴顺序，以匹配之前Tukey HSD的输出
    order = ['普通光照', '黑暗环境', '优化光照']

    for i, metric in enumerate(['总睡眠时间 (小时)', '睡眠效率 (%)', '入睡潜伏期 (分钟)']):
        # 创建箱线图
        df_metrics.boxplot(column=metric, by='光照条件', ax=axes[i], order=order)
        axes[i].set_title(metric)
        axes[i].set_xlabel('光照条件')
        axes[i].set_ylabel(metric)
        axes[i].get_figure().suptitle('')
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    print("睡眠指标箱线图已生成。")

    # 5. 统计检验（ANOVA & Tukey's HSD）
    print("\n--- 正在进行统计检验 ---")
    
    for metric in ['总睡眠时间 (小时)', '睡眠效率 (%)', '入睡潜伏期 (分钟)']:
        print(f"\n--- 对 '{metric}' 进行方差分析 (ANOVA) ---")
        
        group1 = df_metrics[df_metrics['光照条件'] == '普通光照'][metric].dropna()
        group2 = df_metrics[df_metrics['光照条件'] == '黑暗环境'][metric].dropna()
        group3 = df_metrics[df_metrics['光照条件'] == '优化光照'][metric].dropna()
        
        # 移除空值以避免错误
        if len(group1) > 1 and len(group2) > 1 and len(group3) > 1:
            f_statistic, p_value = stats.f_oneway(group1, group2, group3)
            print(f"F-statistic: {f_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")
            
            if p_value < 0.05:
                print("结果解读：P-value 小于显著性水平 0.05，结论：三种光照环境对该指标的影响存在显著差异。")
                print("\n--- 进行事后检验 (Tukey's HSD) ---")
                
                # 事后检验需要扁平化的数据
                data_for_tukey = df_metrics[metric].dropna()
                groups_for_tukey = df_metrics['光照条件'].loc[data_for_tukey.index]
                
                tukey_result = pairwise_tukeyhsd(data_for_tukey, groups_for_tukey, alpha=0.05)
                print(tukey_result)
            else:
                print("结果解读：P-value 大于显著性水平 0.05，结论：三种光照环境对该指标的影响不存在显著差异。")
        else:
            print(f"数据不足，无法对 {metric} 进行统计分析。")

    print("-" * 50)

# 主程序入口
if __name__ == '__main__':
    run_problem3_analysis()
    run_problem4_analysis()
```

```python
# 导入必要的库
import pandas as pd
import numpy as np
from scipy.optimize import nnls
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.interpolate import interp1d

# 设置matplotlib字体以支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

# --- 问题3相关辅助函数 ---
def calculate_rmse(target, predicted):
    """
    计算两个光谱之间的均方根误差。

    参数:
    target (np.array): 目标光谱数据。
    predicted (np.array): 预测光谱数据。

    返回:
    float: 均方根误差。
    """
    return np.sqrt(np.mean((target - predicted)**2))

def calculate_cct(wavelengths, spd):
    """
    计算相关色温 (CCT)。这是一个简化的示例，
    实际计算需要更复杂的算法和CIE标准颜色匹配函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: CCT值。
    """
    # 假设一个简单的加权平均来模拟CCT
    # 实际CCT计算需要XYZ三刺激值，然后转换为CCT
    # 这里用一个简单的加权平均作为占位符
    return np.sum(wavelengths * spd) / np.sum(spd) if np.sum(spd) > 0 else np.nan

def calculate_eml(wavelengths, spd):
    """
    计算等效褪黑激素照度 (EML)。这是一个简化的示例，
    实际计算需要CIE S 026/E:2018标准中定义的黑视蛋白敏感函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: EML值。
    """
    # 假设黑视蛋白敏感性在特定波长处较高
    # 此处使用高斯函数作为占位符
    melanopsin_sensitivity = np.exp(-((wavelengths - 480) ** 2) / (2 * 20 ** 2))
    eml = np.sum(spd * melanopsin_sensitivity)
    return eml

def run_problem3_analysis():
    """
    执行问题3的模型评估和可视化。
    该函数将模型匹配全天太阳光谱的鲁棒性进行分析。
    """
    print("--- 正在进行问题3：模型鲁棒性评估与可视化 ---")
    
    # 1. 读取数据，使用您提供的URL
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_led_spd = pd.read_excel(file_url, sheet_name='Problem 2_LED_SPD')
        df_sun_spd = pd.read_excel(file_url, sheet_name='Problem 3 SUN_SPD')
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 数据预处理
    wavelengths = df_led_spd.iloc[:, 0]
    led_spds = df_led_spd.iloc[:, 1:].to_numpy()
    sun_times = df_sun_spd.columns[1:]
    sun_spds = df_sun_spd.iloc[:, 1:].to_numpy()

    # 确保波长一致性
    if not np.allclose(wavelengths, df_sun_spd.iloc[:, 0]):
        print("警告：LED和太阳光谱的波长不一致，正在进行插值处理...")
        sun_spds_interp = []
        for i in range(sun_spds.shape[1]):
            f = interp1d(df_sun_spd.iloc[:, 0], sun_spds[:, i])
            sun_spds_interp.append(f(wavelengths))
        sun_spds = np.array(sun_spds_interp).T

    # 2. NNLS光谱优化模型
    A = led_spds
    rmse_values = []
    optimal_x_values = []

    for i in range(sun_spds.shape[1]):
        b = sun_spds[:, i]
        x, _ = nnls(A, b)
        optimal_x_values.append(x)
        synthesized_spd = np.dot(A, x)
        rmse = calculate_rmse(b, synthesized_spd)
        rmse_values.append(rmse)

    # 3. 结果可视化：RMSE随时间变化的折线图
    plt.figure(figsize=(10, 6))
    plt.plot(sun_times, rmse_values, marker='o')
    plt.title('模型匹配鲁棒性评估：RMSE随时间的变化')
    plt.xlabel('时间')
    plt.ylabel('均方根误差 (RMSE)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("模型鲁棒性评估图表已生成。")

    # 4. 结果可视化：某一时刻的太阳光谱与合成光谱对比
    midday_index = sun_times.get_loc('12:30:00')
    target_spd = sun_spds[:, midday_index]
    synthesized_spd = np.dot(A, optimal_x_values[midday_index])
    
    plt.figure(figsize=(10, 6))
    plt.plot(wavelengths, target_spd, label=f'目标太阳光谱 ({sun_times[midday_index]})')
    plt.plot(wavelengths, synthesized_spd, '--', label=f'模型合成光谱', linewidth=2)
    plt.title('光谱匹配效果：太阳光谱与合成光谱对比')
    plt.xlabel('波长 (nm)')
    plt.ylabel('相对光强')
    plt.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("光谱匹配效果对比图表已生成。")

    # 5. 扩展评估指标：CCT 和 EML 对比
    cct_sun = calculate_cct(wavelengths, target_spd)
    cct_model = calculate_cct(wavelengths, synthesized_spd)
    eml_sun = calculate_eml(wavelengths, target_spd)
    eml_model = calculate_eml(wavelengths, synthesized_spd)
    
    print("\n--- 扩展评估指标 ---")
    print(f"中午太阳光谱的CCT: {cct_sun:.2f}")
    print(f"模型合成光谱的CCT: {cct_model:.2f}")
    print(f"中午太阳光谱的EML: {eml_sun:.2f}")
    print(f"模型合成光谱的EML: {eml_model:.2f}")

    print("-" * 50)

def run_problem4_analysis():
    """
    执行问题4的睡眠指标分析与可视化，并进行统计检验。
    """
    print("--- 正在进行问题4：睡眠指标分析与可视化 ---")
    
    # 1. 读取数据，使用您提供的URL
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_raw = pd.read_excel(file_url, sheet_name='Problem 4', header=None)
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 2. 数据预处理 - 转换为长格式，方便分析 (基于您的代码)
    subject_header = df_raw.iloc[0].ffill()
    night_header = df_raw.iloc[1]
    new_columns = [f"{sub}_{night}" for sub, night in zip(subject_header, night_header)]
    df_raw.columns = new_columns
    df_sleep_data = df_raw.iloc[2:].copy()
    df_sleep_data = df_sleep_data.apply(pd.to_numeric, errors='coerce')

    metrics_list = []
    epoch_duration_sec = 30
    # 确保光照条件映射与您的代码一致
    condition_mapping = {
        'Night 1': '普通光照',
        'Night 2': '优化光照',
        'Night 3': '黑暗环境'
    }
    
    for col_name in df_sleep_data.columns:
        subject_id, night_str = col_name.split('_')
        condition = condition_mapping.get(night_str, '未知')
        
        sleep_stages = df_sleep_data[col_name].dropna().values
        stages_series = pd.Series(sleep_stages)
        total_epochs = len(stages_series)
        
        # 统计指标
        is_sleep = stages_series.isin([1, 2, 3, 5]) # 假设1,2,3,5为睡眠阶段
        total_bedtime_min = total_epochs * epoch_duration_sec / 60
        
        first_sleep_epoch_idx = stages_series[is_sleep].index[0] if not stages_series[is_sleep].empty else np.nan
        sol = first_sleep_epoch_idx * epoch_duration_sec / 60 if not pd.isna(first_sleep_epoch_idx) else np.nan
        
        tst_epochs = len(stages_series[is_sleep])
        n3_epochs = len(stages_series[stages_series == 3])
        rem_epochs = len(stages_series[stages_series == 5])
        
        tst = tst_epochs * epoch_duration_sec / 60
        se = (tst / total_bedtime_min) * 100 if total_bedtime_min > 0 else 0
        n3_percent = (n3_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        rem_percent = (rem_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        
        awakenings = 0
        if not pd.isna(first_sleep_epoch_idx):
            post_sol_stages = stages_series.iloc[int(first_sleep_epoch_idx):]
            # 计算清醒阶段的连续块数
            awakenings = (post_sol_stages == 4).diff().sum() if (post_sol_stages == 4).any() else 0
            
        metrics_list.append({
            'Subject': subject_id,
            'Condition': condition,
            '总睡眠时间 (小时)': tst / 60, # 转换成小时
            '睡眠效率 (%)': se,
            '入睡潜伏期 (分钟)': sol,
            'N3睡眠比例 (%)': n3_percent,
            'REM睡眠比例 (%)': rem_percent,
            '夜间醒来次数': awakenings
        })
    df_metrics = pd.DataFrame(metrics_list)
    
    # 3. 结果可视化：箱线图
    metrics_to_plot = ['总睡眠时间 (小时)', '睡眠效率 (%)', '入睡潜伏期 (分钟)',
                       'N3睡眠比例 (%)', 'REM睡眠比例 (%)', '夜间醒来次数']
                       
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('不同光照条件下睡眠指标的分布', fontsize=16)
    
    order = ['普通光照', '优化光照', '黑暗环境']

    for i, metric in enumerate(metrics_to_plot):
        ax = axes.flatten()[i]
        df_metrics.boxplot(column=metric, by='Condition', ax=ax, order=order, grid=False, rot=0)
        ax.set_title(metric, fontsize=12)
        ax.set_xlabel('光照条件', fontsize=10)
        ax.set_ylabel(metric, fontsize=10)
        ax.get_figure().suptitle('') # 移除子图自带的suptitle
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    print("睡眠指标箱线图已生成。")

    # 4. 统计检验（ANOVA & Tukey's HSD）
    print("\n--- 正在进行统计检验 ---")
    for metric in metrics_to_plot:
        print(f"\n--- 对 '{metric}' 进行方差分析 (ANOVA) ---")
        
        group1 = df_metrics[df_metrics['Condition'] == '普通光照'][metric].dropna()
        group2 = df_metrics[df_metrics['Condition'] == '优化光照'][metric].dropna()
        group3 = df_metrics[df_metrics['Condition'] == '黑暗环境'][metric].dropna()
        
        if len(group1) > 1 and len(group2) > 1 and len(group3) > 1:
            f_statistic, p_value = stats.f_oneway(group1, group2, group3)
            print(f"F-statistic: {f_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")
            
            if p_value < 0.05:
                print("结果解读：P-value 小于显著性水平 0.05，结论：三种光照环境对该指标的影响存在显著差异。")
                print("\n--- 进行事后检验 (Tukey's HSD) ---")
                
                data_for_tukey = df_metrics[metric].dropna()
                groups_for_tukey = df_metrics['Condition'].loc[data_for_tukey.index]
                
                tukey_result = pairwise_tukeyhsd(data_for_tukey, groups_for_tukey, alpha=0.05)
                print(tukey_result)
            else:
                print("结果解读：P-value 大于显著性水平 0.05，结论：三种光照环境对该指标的影响不存在显著差异。")
        else:
            print(f"数据不足，无法对 {metric} 进行统计分析。")

    print("-" * 50)

# 主程序入口
if __name__ == '__main__':
    run_problem3_analysis()
    run_problem4_analysis()
```

```python
 导入必要的库
import pandas as pd
import numpy as np
from scipy.optimize import nnls
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.interpolate import interp1d

# 设置matplotlib字体以支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

# --- 问题3相关辅助函数 ---
def calculate_rmse(target, predicted):
    """
    计算两个光谱之间的均方根误差。

    参数:
    target (np.array): 目标光谱数据。
    predicted (np.array): 预测光谱数据。

    返回:
    float: 均方根误差。
    """
    return np.sqrt(np.mean((target - predicted)**2))

def calculate_cct(wavelengths, spd):
    """
    计算相关色温 (CCT)。这是一个简化的示例，
    实际计算需要更复杂的算法和CIE标准颜色匹配函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: CCT值。
    """
    # 假设一个简单的加权平均来模拟CCT
    # 实际CCT计算需要XYZ三刺激值，然后转换为CCT
    # 这里用一个简单的加权平均作为占位符
    return np.sum(wavelengths * spd) / np.sum(spd) if np.sum(spd) > 0 else np.nan

def calculate_eml(wavelengths, spd):
    """
    计算等效褪黑激素照度 (EML)。这是一个简化的示例，
    实际计算需要CIE S 026/E:2018标准中定义的黑视蛋白敏感函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: EML值。
    """
    # 假设黑视蛋白敏感性在特定波长处较高
    # 此处使用高斯函数作为占位符
    melanopsin_sensitivity = np.exp(-((wavelengths - 480) ** 2) / (2 * 20 ** 2))
    eml = np.sum(spd * melanopsin_sensitivity)
    return eml

def run_problem3_analysis():
    """
    执行问题3的模型评估和可视化。
    该函数将模型匹配全天太阳光谱的鲁棒性进行分析。
    """
    print("--- 正在进行问题3：模型鲁棒性评估与可视化 ---")
    
    # 1. 读取数据，使用您提供的URL
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        # 读取数据时，不指定 header，以便我们手动处理
        df_led_spd = pd.read_excel(file_url, sheet_name='Problem 2_LED_SPD')
        df_sun_spd = pd.read_excel(file_url, sheet_name='Problem 3 SUN_SPD')
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 数据预处理
    # 清洗波长列，去除单位并转换为数值
    wavelengths_raw = df_led_spd.iloc[:, 0]
    wavelengths = wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    led_spds = df_led_spd.iloc[:, 1:].to_numpy()
    
    # 清洗太阳光谱的波长列
    sun_wavelengths_raw = df_sun_spd.iloc[:, 0]
    sun_wavelengths = sun_wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    sun_times = df_sun_spd.columns[1:]
    sun_spds = df_sun_spd.iloc[:, 1:].to_numpy()

    # 确保波长一致性
    if not np.allclose(wavelengths, sun_wavelengths):
        print("警告：LED和太阳光谱的波长不一致，正在进行插值处理...")
        sun_spds_interp = []
        for i in range(sun_spds.shape[1]):
            f = interp1d(sun_wavelengths, sun_spds[:, i])
            sun_spds_interp.append(f(wavelengths))
        sun_spds = np.array(sun_spds_interp).T

    # 2. NNLS光谱优化模型
    A = led_spds
    rmse_values = []
    optimal_x_values = []
    cct_sun_values = []
    cct_model_values = []
    eml_sun_values = []
    eml_model_values = []

    for i in range(sun_spds.shape[1]):
        b = sun_spds[:, i]
        x, _ = nnls(A, b)
        optimal_x_values.append(x)
        synthesized_spd = np.dot(A, x)
        
        # 计算评估指标
        rmse = calculate_rmse(b, synthesized_spd)
        rmse_values.append(rmse)
        
        cct_sun = calculate_cct(wavelengths, b)
        cct_model = calculate_cct(wavelengths, synthesized_spd)
        cct_sun_values.append(cct_sun)
        cct_model_values.append(cct_model)
        
        eml_sun = calculate_eml(wavelengths, b)
        eml_model = calculate_eml(wavelengths, synthesized_spd)
        eml_sun_values.append(eml_sun)
        eml_model_values.append(eml_model)

    # 3. 结果可视化：RMSE随时间变化的折线图
    plt.figure(figsize=(10, 6))
    plt.plot(sun_times, rmse_values, marker='o')
    plt.title('模型匹配鲁棒性评估：RMSE随时间的变化')
    plt.xlabel('时间')
    plt.ylabel('均方根误差 (RMSE)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("模型鲁棒性评估图表已生成。")

    # 4. 结果可视化：CCT 和 EML 随时间变化的对比图
    fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)
    
    axes[0].plot(sun_times, cct_sun_values, marker='o', label='太阳光谱CCT')
    axes[0].plot(sun_times, cct_model_values, marker='x', linestyle='--', label='模型合成光谱CCT')
    axes[0].set_title('相关色温 (CCT) 随时间的变化')
    axes[0].set_ylabel('CCT')
    axes[0].legend()
    axes[0].grid(True, linestyle='--', alpha=0.6)

    axes[1].plot(sun_times, eml_sun_values, marker='o', label='太阳光谱EML')
    axes[1].plot(sun_times, eml_model_values, marker='x', linestyle='--', label='模型合成光谱EML')
    axes[1].set_title('等效褪黑激素照度 (EML) 随时间的变化')
    axes[1].set_xlabel('时间')
    axes[1].set_ylabel('EML')
    axes[1].legend()
    axes[1].grid(True, linestyle='--', alpha=0.6)
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("CCT和EML对比图表已生成。")
    print("-" * 50)

def run_problem4_analysis():
    """
    执行问题4的睡眠指标分析与可视化，并进行统计检验。
    """
    print("--- 正在进行问题4：睡眠指标分析与可视化 ---")
    
    # 1. 读取数据，使用您提供的URL
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_raw = pd.read_excel(file_url, sheet_name='Problem 4', header=None)
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 2. 数据预处理 - 转换为长格式，方便分析 (基于您的代码)
    subject_header = df_raw.iloc[0].ffill()
    night_header = df_raw.iloc[1]
    new_columns = [f"{sub}_{night}" for sub, night in zip(subject_header, night_header)]
    df_raw.columns = new_columns
    df_sleep_data = df_raw.iloc[2:].copy()
    df_sleep_data = df_sleep_data.apply(pd.to_numeric, errors='coerce')

    metrics_list = []
    epoch_duration_sec = 30
    condition_mapping = {
        'Night 1': '普通光照',
        'Night 2': '黑暗环境', # 假设优化光照效果最好
        'Night 3': '优化光照'
    }
    
    for col_name in df_sleep_data.columns:
        subject_id, night_str = col_name.split('_')
        condition = condition_mapping.get(night_str, '未知')
        
        sleep_stages = df_sleep_data[col_name].dropna().values
        stages_series = pd.Series(sleep_stages)
        total_epochs = len(stages_series)
        
        # 统计指标
        is_sleep = stages_series.isin([1, 2, 3, 5]) # 假设1,2,3,5为睡眠阶段
        total_bedtime_min = total_epochs * epoch_duration_sec / 60
        
        first_sleep_epoch_idx = stages_series[is_sleep].index[0] if not stages_series[is_sleep].empty else np.nan
        sol = (first_sleep_epoch_idx - stages_series.index[0]) * epoch_duration_sec / 60 if not pd.isna(first_sleep_epoch_idx) else np.nan
        
        tst_epochs = len(stages_series[is_sleep])
        n3_epochs = len(stages_series[stages_series == 3])
        rem_epochs = len(stages_series[stages_series == 5])
        
        tst_min = tst_epochs * epoch_duration_sec / 60
        se = (tst_epochs / total_epochs) * 100 if total_epochs > 0 else 0
        n3_percent = (n3_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        rem_percent = (rem_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        
        awakenings = 0
        if not pd.isna(first_sleep_epoch_idx):
            post_sol_stages = stages_series.iloc[int(first_sleep_epoch_idx):]
            awakenings = (post_sol_stages == 4).diff().sum() if (post_sol_stages == 4).any() else 0
            
        metrics_list.append({
            'Subject': subject_id,
            'Condition': condition,
            '总睡眠时间 (分钟)': tst_min,
            '睡眠效率 (%)': se,
            '入睡潜伏期 (分钟)': sol,
            'N3睡眠比例 (%)': n3_percent,
            'REM睡眠比例 (%)': rem_percent,
            '夜间醒来次数': awakenings
        })
    df_metrics = pd.DataFrame(metrics_list)
    
    # 3. 结果可视化：箱线图
    metrics_to_plot = ['总睡眠时间 (分钟)', '睡眠效率 (%)', '入睡潜伏期 (分钟)',
                       'N3睡眠比例 (%)', 'REM睡眠比例 (%)', '夜间醒来次数']
                       
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('不同光照条件下睡眠指标的分布', fontsize=16)
    
    order = ['普通光照', '黑暗环境', '优化光照']

    for i, metric in enumerate(metrics_to_plot):
        ax = axes.flatten()[i]
        df_metrics.boxplot(column=metric, by='Condition', ax=ax, order=order, grid=False, rot=0)
        ax.set_title(metric, fontsize=12)
        ax.set_xlabel('光照条件', fontsize=10)
        ax.set_ylabel(metric, fontsize=10)
        ax.get_figure().suptitle('') # 移除子图自带的suptitle
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    print("睡眠指标箱线图已生成。")

    # 4. 统计检验（ANOVA & Tukey's HSD）
    print("\n--- 正在进行统计检验 ---")
    for metric in metrics_to_plot:
        print(f"\n--- 对 '{metric}' 进行方差分析 (ANOVA) ---")
        
        group1 = df_metrics[df_metrics['Condition'] == '普通光照'][metric].dropna()
        group2 = df_metrics[df_metrics['Condition'] == '黑暗环境'][metric].dropna()
        group3 = df_metrics[df_metrics['Condition'] == '优化光照'][metric].dropna()
        
        if len(group1) > 1 and len(group2) > 1 and len(group3) > 1:
            f_statistic, p_value = stats.f_oneway(group1, group2, group3)
            print(f"F-statistic: {f_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")
            
            if p_value < 0.05:
                print("结果解读：P-value 小于显著性水平 0.05，结论：三种光照环境对该指标的影响存在显著差异。")
                print("\n--- 进行事后检验 (Tukey's HSD) ---")
                
                data_for_tukey = df_metrics[metric].dropna()
                groups_for_tukey = df_metrics['Condition'].loc[data_for_tukey.index]
                
                tukey_result = pairwise_tukeyhsd(data_for_tukey, groups_for_tukey, alpha=0.05)
                print(tukey_result)
            else:
                print("结果解读：P-value 大于显著性水平 0.05，结论：三种光照环境对该指标的影响不存在显著差异。")
        else:
            print(f"数据不足，无法对 {metric} 进行统计分析。")

    print("-" * 50)

# 主程序入口
if __name__ == '__main__':
    run_problem3_analysis()
    run_problem4_analysis()
```

```python
import pandas as pd
import numpy as np
from scipy.optimize import nnls
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.interpolate import interp1d

# 设置matplotlib字体以支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

# --- 问题3相关辅助函数 ---
def calculate_rmse(target, predicted):
    """
    计算两个光谱之间的均方根误差。

    参数:
    target (np.array): 目标光谱数据。
    predicted (np.array): 预测光谱数据。

    返回:
    float: 均方根误差。
    """
    return np.sqrt(np.mean((target - predicted)**2))

def calculate_cct(wavelengths, spd):
    """
    计算相关色温 (CCT)。这是一个简化的示例，
    实际计算需要更复杂的算法和CIE标准颜色匹配函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: CCT值。
    """
    # 假设一个简单的加权平均来模拟CCT
    # 实际CCT计算需要XYZ三刺激值，然后转换为CCT
    # 这里用一个简单的加权平均作为占位符
    return np.sum(wavelengths * spd) / np.sum(spd) if np.sum(spd) > 0 else np.nan

def calculate_eml(wavelengths, spd):
    """
    计算等效褪黑激素照度 (EML)。这是一个简化的示例，
    实际计算需要CIE S 026/E:2018标准中定义的黑视蛋白敏感函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: EML值。
    """
    # 假设黑视蛋白敏感性在特定波长处较高
    # 此处使用高斯函数作为占位符
    melanopsin_sensitivity = np.exp(-((wavelengths - 480) ** 2) / (2 * 20 ** 2))
    eml = np.sum(spd * melanopsin_sensitivity)
    return eml

def run_problem3_analysis():
    """
    执行问题3的模型评估和可视化。
    该函数将模型匹配全天太阳光谱的鲁棒性进行分析。
    """
    print("--- 正在进行问题3：模型鲁棒性评估与可视化 ---")
    
    # 1. 读取数据
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_led_spd = pd.read_excel(file_url, sheet_name='Problem 2_LED_SPD')
        df_sun_spd = pd.read_excel(file_url, sheet_name='Problem 3 SUN_SPD')
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 数据预处理
    # 清洗波长列，去除单位并转换为数值
    wavelengths_raw = df_led_spd.iloc[:, 0]
    wavelengths = wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    led_spds = df_led_spd.iloc[:, 1:].to_numpy()
    
    # 清洗太阳光谱的波长列
    sun_wavelengths_raw = df_sun_spd.iloc[:, 0]
    sun_wavelengths = sun_wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    
    # 修复：将时间列名转换为字符串，以确保绘图时不会出现类型错误
    sun_times = [str(t) for t in df_sun_spd.columns[1:]]
    sun_spds = df_sun_spd.iloc[:, 1:].to_numpy()

    # 确保波长一致性
    if not np.allclose(wavelengths, sun_wavelengths):
        print("警告：LED和太阳光谱的波长不一致，正在进行插值处理...")
        sun_spds_interp = []
        for i in range(sun_spds.shape[1]):
            f = interp1d(sun_wavelengths, sun_spds[:, i])
            sun_spds_interp.append(f(wavelengths))
        sun_spds = np.array(sun_spds_interp).T

    # 2. NNLS光谱优化模型
    A = led_spds
    rmse_values = []
    optimal_x_values = []
    cct_sun_values = []
    cct_model_values = []
    eml_sun_values = []
    eml_model_values = []

    for i in range(sun_spds.shape[1]):
        b = sun_spds[:, i]
        x, _ = nnls(A, b)
        optimal_x_values.append(x)
        synthesized_spd = np.dot(A, x)
        
        # 计算评估指标
        rmse = calculate_rmse(b, synthesized_spd)
        rmse_values.append(rmse)
        
        cct_sun = calculate_cct(wavelengths, b)
        cct_model = calculate_cct(wavelengths, synthesized_spd)
        cct_sun_values.append(cct_sun)
        cct_model_values.append(cct_model)
        
        eml_sun = calculate_eml(wavelengths, b)
        eml_model = calculate_eml(wavelengths, synthesized_spd)
        eml_sun_values.append(eml_sun)
        eml_model_values.append(eml_model)

    # 3. 结果可视化：RMSE随时间变化的折线图
    plt.figure(figsize=(10, 6))
    plt.plot(sun_times, rmse_values, marker='o')
    plt.title('模型匹配鲁棒性评估：RMSE随时间的变化')
    plt.xlabel('时间')
    plt.ylabel('均方根误差 (RMSE)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("模型鲁棒性评估图表已生成。")

    # 4. 结果可视化：CCT 和 EML 随时间变化的对比图
    fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)
    
    axes[0].plot(sun_times, cct_sun_values, marker='o', label='太阳光谱CCT')
    axes[0].plot(sun_times, cct_model_values, marker='x', linestyle='--', label='模型合成光谱CCT')
    axes[0].set_title('相关色温 (CCT) 随时间的变化')
    axes[0].set_ylabel('CCT')
    axes[0].legend()
    axes[0].grid(True, linestyle='--', alpha=0.6)

    axes[1].plot(sun_times, eml_sun_values, marker='o', label='太阳光谱EML')
    axes[1].plot(sun_times, eml_model_values, marker='x', linestyle='--', label='模型合成光谱EML')
    axes[1].set_title('等效褪黑激素照度 (EML) 随时间的变化')
    axes[1].set_xlabel('时间')
    axes[1].set_ylabel('EML')
    axes[1].legend()
    axes[1].grid(True, linestyle='--', alpha=0.6)
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("CCT和EML对比图表已生成。")
    print("-" * 50)

def run_problem4_analysis():
    """
    执行问题4的睡眠指标分析与可视化，并进行统计检验。
    """
    print("--- 正在进行问题4：睡眠指标分析与可视化 ---")
    
    # 1. 读取数据
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_raw = pd.read_excel(file_url, sheet_name='Problem 4', header=None)
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 2. 数据预处理 - 转换为长格式，方便分析 (基于您的代码)
    subject_header = df_raw.iloc[0].ffill()
    night_header = df_raw.iloc[1]
    new_columns = [f"{sub}_{night}" for sub, night in zip(subject_header, night_header)]
    df_raw.columns = new_columns
    df_sleep_data = df_raw.iloc[2:].copy()
    df_sleep_data = df_sleep_data.apply(pd.to_numeric, errors='coerce')

    metrics_list = []
    epoch_duration_sec = 30
    condition_mapping = {
        'Night 1': '普通光照',
        'Night 2': '优化光照',
        'Night 3': '黑暗环境'
    }
    
    for col_name in df_sleep_data.columns:
        subject_id, night_str = col_name.split('_')
        condition = condition_mapping.get(night_str, '未知')
        
        sleep_stages = df_sleep_data[col_name].dropna().values
        stages_series = pd.Series(sleep_stages)
        total_epochs = len(stages_series)
        
        # 统计指标
        is_sleep = stages_series.isin([1, 2, 3, 5]) # 假设1,2,3,5为睡眠阶段
        total_bedtime_min = total_epochs * epoch_duration_sec / 60
        
        first_sleep_epoch_idx = stages_series[is_sleep].index[0] if not stages_series[is_sleep].empty else np.nan
        sol = (first_sleep_epoch_idx - stages_series.index[0]) * epoch_duration_sec / 60 if not pd.isna(first_sleep_epoch_idx) else np.nan
        
        tst_epochs = len(stages_series[is_sleep])
        n3_epochs = len(stages_series[stages_series == 3])
        rem_epochs = len(stages_series[stages_series == 5])
        
        tst_min = tst_epochs * epoch_duration_sec / 60
        se = (tst_epochs / total_epochs) * 100 if total_epochs > 0 else 0
        n3_percent = (n3_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        rem_percent = (rem_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        
        awakenings = 0
        if not pd.isna(first_sleep_epoch_idx):
            post_sol_stages = stages_series.iloc[int(first_sleep_epoch_idx):]
            awakenings = (post_sol_stages == 4).diff().sum() if (post_sol_stages == 4).any() else 0
            
        metrics_list.append({
            'Subject': subject_id,
            'Condition': condition,
            '总睡眠时间 (分钟)': tst_min,
            '睡眠效率 (%)': se,
            '入睡潜伏期 (分钟)': sol,
            'N3睡眠比例 (%)': n3_percent,
            'REM睡眠比例 (%)': rem_percent,
            '夜间醒来次数': awakenings
        })
    df_metrics = pd.DataFrame(metrics_list)
    
    # 3. 结果可视化：箱线图
    metrics_to_plot = ['总睡眠时间 (分钟)', '睡眠效率 (%)', '入睡潜伏期 (分钟)',
                       'N3睡眠比例 (%)', 'REM睡眠比例 (%)', '夜间醒来次数']
                       
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('不同光照条件下睡眠指标的分布', fontsize=16)
    
    order = ['普通光照', '优化光照', '黑暗环境']

    for i, metric in enumerate(metrics_to_plot):
        ax = axes.flatten()[i]
        df_metrics.boxplot(column=metric, by='Condition', ax=ax, order=order, grid=False, rot=0)
        ax.set_title(metric, fontsize=12)
        ax.set_xlabel('光照条件', fontsize=10)
        ax.set_ylabel(metric, fontsize=10)
        ax.get_figure().suptitle('') # 移除子图自带的suptitle
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    print("睡眠指标箱线图已生成。")

    # 4. 统计检验（ANOVA & Tukey's HSD）
    print("\n--- 正在进行统计检验 ---")
    for metric in metrics_to_plot:
        print(f"\n--- 对 '{metric}' 进行方差分析 (ANOVA) ---")
        
        group1 = df_metrics[df_metrics['Condition'] == '普通光照'][metric].dropna()
        group2 = df_metrics[df_metrics['Condition'] == '优化光照'][metric].dropna()
        group3 = df_metrics[df_metrics['Condition'] == '黑暗环境'][metric].dropna()
        
        if len(group1) > 1 and len(group2) > 1 and len(group3) > 1:
            f_statistic, p_value = stats.f_oneway(group1, group2, group3)
            print(f"F-statistic: {f_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")
            
            if p_value < 0.05:
                print("结果解读：P-value 小于显著性水平 0.05，结论：三种光照环境对该指标的影响存在显著差异。")
                print("\n--- 进行事后检验 (Tukey's HSD) ---")
                
                data_for_tukey = df_metrics[metric].dropna()
                groups_for_tukey = df_metrics['Condition'].loc[data_for_tukey.index]
                
                tukey_result = pairwise_tukeyhsd(data_for_tukey, groups_for_tukey, alpha=0.05)
                print(tukey_result)
            else:
                print("结果解读：P-value 大于显著性水平 0.05，结论：三种光照环境对该指标的影响不存在显著差异。")
        else:
            print(f"数据不足，无法对 {metric} 进行统计分析。")

    print("-" * 50)

# 主程序入口
if __name__ == '__main__':
    run_problem3_analysis()
    run_problem4_analysis()
```

```python
# 导入必要的库
import pandas as pd
import numpy as np
from scipy.optimize import nnls
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.interpolate import interp1d

# 设置matplotlib字体以支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

# --- 问题3相关辅助函数 ---
def calculate_rmse(target, predicted):
    """
    计算两个光谱之间的均方根误差。

    参数:
    target (np.array): 目标光谱数据。
    predicted (np.array): 预测光谱数据。

    返回:
    float: 均方根误差。
    """
    return np.sqrt(np.mean((target - predicted)**2))

def calculate_cct(wavelengths, spd):
    """
    计算相关色温 (CCT)。这是一个简化的示例，
    实际计算需要更复杂的算法和CIE标准颜色匹配函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: CCT值。
    """
    # 假设一个简单的加权平均来模拟CCT
    # 实际CCT计算需要XYZ三刺激值，然后转换为CCT
    # 这里用一个简单的加权平均作为占位符
    return np.sum(wavelengths * spd) / np.sum(spd) if np.sum(spd) > 0 else np.nan

def calculate_eml(wavelengths, spd):
    """
    计算等效褪黑激素照度 (EML)。这是一个简化的示例，
    实际计算需要CIE S 026/E:2018标准中定义的黑视蛋白敏感函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: EML值。
    """
    # 假设黑视蛋白敏感性在特定波长处较高
    # 此处使用高斯函数作为占位符
    melanopsin_sensitivity = np.exp(-((wavelengths - 480) ** 2) / (2 * 20 ** 2))
    eml = np.sum(spd * melanopsin_sensitivity)
    return eml

def run_problem3_analysis():
    """
    执行问题3的模型评估和可视化。
    该函数将模型匹配全天太阳光谱的鲁棒性进行分析。
    """
    print("--- 正在进行问题3：模型鲁棒性评估与可视化 ---")
    
    # 1. 读取数据
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_led_spd = pd.read_excel(file_url, sheet_name='Problem 2_LED_SPD')
        df_sun_spd = pd.read_excel(file_url, sheet_name='Problem 3 SUN_SPD')
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 数据预处理
    # 清洗波长列，去除单位并转换为数值
    wavelengths_raw = df_led_spd.iloc[:, 0]
    wavelengths = wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    led_spds = df_led_spd.iloc[:, 1:].to_numpy()
    
    # 清洗太阳光谱的波长列
    sun_wavelengths_raw = df_sun_spd.iloc[:, 0]
    sun_wavelengths = sun_wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    
    # 修复：将时间列名转换为字符串，以确保绘图时不会出现类型错误
    sun_times = [str(t) for t in df_sun_spd.columns[1:]]
    sun_spds = df_sun_spd.iloc[:, 1:].to_numpy()

    # 确保波长一致性
    if not np.allclose(wavelengths, sun_wavelengths):
        print("警告：LED和太阳光谱的波长不一致，正在进行插值处理...")
        sun_spds_interp = []
        for i in range(sun_spds.shape[1]):
            f = interp1d(sun_wavelengths, sun_spds[:, i])
            sun_spds_interp.append(f(wavelengths))
        sun_spds = np.array(sun_spds_interp).T

    # 2. NNLS光谱优化模型
    A = led_spds
    rmse_values = []
    optimal_x_values = []
    cct_sun_values = []
    cct_model_values = []
    eml_sun_values = []
    eml_model_values = []

    for i in range(sun_spds.shape[1]):
        b = sun_spds[:, i]
        x, _ = nnls(A, b)
        optimal_x_values.append(x)
        synthesized_spd = np.dot(A, x)
        
        # 计算评估指标
        rmse = calculate_rmse(b, synthesized_spd)
        rmse_values.append(rmse)
        
        cct_sun = calculate_cct(wavelengths, b)
        cct_model = calculate_cct(wavelengths, synthesized_spd)
        cct_sun_values.append(cct_sun)
        cct_model_values.append(cct_model)
        
        eml_sun = calculate_eml(wavelengths, b)
        eml_model = calculate_eml(wavelengths, synthesized_spd)
        eml_sun_values.append(eml_sun)
        eml_model_values.append(eml_model)

    # 3. 结果可视化：RMSE随时间变化的折线图
    plt.figure(figsize=(10, 6))
    plt.plot(sun_times, rmse_values, marker='o')
    plt.title('模型匹配鲁棒性评估：RMSE随时间的变化')
    plt.xlabel('时间')
    plt.ylabel('均方根误差 (RMSE)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("模型鲁棒性评估图表已生成。")

    # 4. 结果可视化：CCT 和 EML 随时间变化的对比图
    fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)
    
    axes[0].plot(sun_times, cct_sun_values, marker='o', label='太阳光谱CCT')
    axes[0].plot(sun_times, cct_model_values, marker='x', linestyle='--', label='模型合成光谱CCT')
    axes[0].set_title('相关色温 (CCT) 随时间的变化')
    axes[0].set_ylabel('CCT')
    axes[0].legend()
    axes[0].grid(True, linestyle='--', alpha=0.6)

    axes[1].plot(sun_times, eml_sun_values, marker='o', label='太阳光谱EML')
    axes[1].plot(sun_times, eml_model_values, marker='x', linestyle='--', label='模型合成光谱EML')
    axes[1].set_title('等效褪黑激素照度 (EML) 随时间的变化')
    axes[1].set_xlabel('时间')
    axes[1].set_ylabel('EML')
    axes[1].legend()
    axes[1].grid(True, linestyle='--', alpha=0.6)
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("CCT和EML对比图表已生成。")
    print("-" * 50)

def run_problem4_analysis():
    """
    执行问题4的睡眠指标分析与可视化，并进行统计检验。
    """
    print("--- 正在进行问题4：睡眠指标分析与可视化 ---")
    
    # 1. 读取数据
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_raw = pd.read_excel(file_url, sheet_name='Problem 4', header=None)
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 2. 数据预处理 - 转换为长格式，方便分析 (基于您的代码)
    subject_header = df_raw.iloc[0].ffill()
    night_header = df_raw.iloc[1]
    new_columns = [f"{sub}_{night}" for sub, night in zip(subject_header, night_header)]
    df_raw.columns = new_columns
    df_sleep_data = df_raw.iloc[2:].copy()
    df_sleep_data = df_sleep_data.apply(pd.to_numeric, errors='coerce')

    metrics_list = []
    epoch_duration_sec = 30
    condition_mapping = {
        'Night 1': '普通光照',
        'Night 2': '优化光照',
        'Night 3': '黑暗环境'
    }
    
    for col_name in df_sleep_data.columns:
        subject_id, night_str = col_name.split('_')
        condition = condition_mapping.get(night_str, '未知')
        
        sleep_stages = df_sleep_data[col_name].dropna().values
        stages_series = pd.Series(sleep_stages)
        total_epochs = len(stages_series)
        
        # 统计指标
        is_sleep = stages_series.isin([1, 2, 3, 5]) # 假设1,2,3,5为睡眠阶段
        total_bedtime_min = total_epochs * epoch_duration_sec / 60
        
        first_sleep_epoch_idx = stages_series[is_sleep].index[0] if not stages_series[is_sleep].empty else np.nan
        sol = (first_sleep_epoch_idx - stages_series.index[0]) * epoch_duration_sec / 60 if not pd.isna(first_sleep_epoch_idx) else np.nan
        
        tst_epochs = len(stages_series[is_sleep])
        n3_epochs = len(stages_series[stages_series == 3])
        rem_epochs = len(stages_series[stages_series == 5])
        
        tst_min = tst_epochs * epoch_duration_sec / 60
        se = (tst_epochs / total_epochs) * 100 if total_epochs > 0 else 0
        n3_percent = (n3_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        rem_percent = (rem_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        
        awakenings = 0
        if not pd.isna(first_sleep_epoch_idx):
            post_sol_stages = stages_series.iloc[int(first_sleep_epoch_idx):]
            awakenings = (post_sol_stages == 4).diff().sum() if (post_sol_stages == 4).any() else 0
            
        metrics_list.append({
            'Subject': subject_id,
            'Condition': condition,
            '总睡眠时间 (分钟)': tst_min,
            '睡眠效率 (%)': se,
            '入睡潜伏期 (分钟)': sol,
            'N3睡眠比例 (%)': n3_percent,
            'REM睡眠比例 (%)': rem_percent,
            '夜间醒来次数': awakenings
        })
    df_metrics = pd.DataFrame(metrics_list)
    
    # 修复：将Condition列转换为有序的分类类型，以确保箱线图的顺序
    order = ['普通光照', '优化光照', '黑暗环境']
    df_metrics['Condition'] = pd.Categorical(df_metrics['Condition'], categories=order, ordered=True)
    df_metrics.sort_values('Condition', inplace=True)

    # 3. 结果可视化：箱线图
    metrics_to_plot = ['总睡眠时间 (分钟)', '睡眠效率 (%)', '入睡潜伏期 (分钟)',
                       'N3睡眠比例 (%)', 'REM睡眠比例 (%)', '夜间醒来次数']
                       
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('不同光照条件下睡眠指标的分布', fontsize=16)

    for i, metric in enumerate(metrics_to_plot):
        ax = axes.flatten()[i]
        df_metrics.boxplot(column=metric, by='Condition', ax=ax, grid=False, rot=0)
        ax.set_title(metric, fontsize=12)
        ax.set_xlabel('光照条件', fontsize=10)
        ax.set_ylabel(metric, fontsize=10)
        ax.get_figure().suptitle('') # 移除子图自带的suptitle
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    print("睡眠指标箱线图已生成。")

    # 4. 统计检验（ANOVA & Tukey's HSD）
    print("\n--- 正在进行统计检验 ---")
    for metric in metrics_to_plot:
        print(f"\n--- 对 '{metric}' 进行方差分析 (ANOVA) ---")
        
        group1 = df_metrics[df_metrics['Condition'] == '普通光照'][metric].dropna()
        group2 = df_metrics[df_metrics['Condition'] == '优化光照'][metric].dropna()
        group3 = df_metrics[df_metrics['Condition'] == '黑暗环境'][metric].dropna()
        
        if len(group1) > 1 and len(group2) > 1 and len(group3) > 1:
            f_statistic, p_value = stats.f_oneway(group1, group2, group3)
            print(f"F-statistic: {f_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")
            
            if p_value < 0.05:
                print("结果解读：P-value 小于显著性水平 0.05，结论：三种光照环境对该指标的影响存在显著差异。")
                print("\n--- 进行事后检验 (Tukey's HSD) ---")
                
                data_for_tukey = df_metrics[metric].dropna()
                groups_for_tukey = df_metrics['Condition'].loc[data_for_tukey.index]
                
                tukey_result = pairwise_tukeyhsd(data_for_tukey, groups_for_tukey, alpha=0.05)
                print(tukey_result)
            else:
                print("结果解读：P-value 大于显著性水平 0.05，结论：三种光照环境对该指标的影响不存在显著差异。")
        else:
            print(f"数据不足，无法对 {metric} 进行统计分析。")

    print("-" * 50)

# 主程序入口
if __name__ == '__main__':
    run_problem3_analysis()
    run_problem4_analysis()
```

```python
# 导入必要的库
import pandas as pd
import numpy as np
from scipy.optimize import nnls
from scipy import stats
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
from scipy.interpolate import interp1d

# 设置matplotlib字体以支持中文显示
plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体
plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像时负号'-'显示为方块的问题

# --- 问题3相关辅助函数 ---
def calculate_rmse(target, predicted):
    """
    计算两个光谱之间的均方根误差。

    参数:
    target (np.array): 目标光谱数据。
    predicted (np.array): 预测光谱数据。

    返回:
    float: 均方根误差。
    """
    return np.sqrt(np.mean((target - predicted)**2))

def calculate_cct(wavelengths, spd):
    """
    计算相关色温 (CCT)。这是一个简化的示例，
    实际计算需要更复杂的算法和CIE标准颜色匹配函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: CCT值。
    """
    # 假设一个简单的加权平均来模拟CCT
    # 实际CCT计算需要XYZ三刺激值，然后转换为CCT
    # 这里用一个简单的加权平均作为占位符
    return np.sum(wavelengths * spd) / np.sum(spd) if np.sum(spd) > 0 else np.nan

def calculate_eml(wavelengths, spd):
    """
    计算等效褪黑激素照度 (EML)。这是一个简化的示例，
    实际计算需要CIE S 026/E:2018标准中定义的黑视蛋白敏感函数。
    此函数仅用于演示目的，其结果不应作为精确值。
    
    参数:
    wavelengths (np.array): 波长。
    spd (np.array): 光谱功率分布。

    返回:
    float: EML值。
    """
    # 假设黑视蛋白敏感性在特定波长处较高
    # 此处使用高斯函数作为占位符
    melanopsin_sensitivity = np.exp(-((wavelengths - 480) ** 2) / (2 * 20 ** 2))
    eml = np.sum(spd * melanopsin_sensitivity)
    return eml

def run_problem3_analysis():
    """
    执行问题3的模型评估和可视化。
    该函数将模型匹配全天太阳光谱的鲁棒性进行分析。
    """
    print("--- 正在进行问题3：模型鲁棒性评估与可视化 ---")
    
    # 1. 读取数据
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_led_spd = pd.read_excel(file_url, sheet_name='Problem 2_LED_SPD')
        df_sun_spd = pd.read_excel(file_url, sheet_name='Problem 3 SUN_SPD')
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 数据预处理
    # 清洗波长列，去除单位并转换为数值
    wavelengths_raw = df_led_spd.iloc[:, 0]
    wavelengths = wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    led_spds = df_led_spd.iloc[:, 1:].to_numpy()
    
    # 清洗太阳光谱的波长列
    sun_wavelengths_raw = df_sun_spd.iloc[:, 0]
    sun_wavelengths = sun_wavelengths_raw.astype(str).str.replace(r'\(.*\)', '', regex=True).astype(float)
    
    # 将时间列名转换为字符串，以确保绘图时不会出现类型错误
    sun_times = [str(t) for t in df_sun_spd.columns[1:]]
    sun_spds = df_sun_spd.iloc[:, 1:].to_numpy()

    # 确保波长一致性
    if not np.allclose(wavelengths, sun_wavelengths):
        print("警告：LED和太阳光谱的波长不一致，正在进行插值处理...")
        sun_spds_interp = []
        for i in range(sun_spds.shape[1]):
            f = interp1d(sun_wavelengths, sun_spds[:, i])
            sun_spds_interp.append(f(wavelengths))
        sun_spds = np.array(sun_spds_interp).T

    # 2. NNLS光谱优化模型
    A = led_spds
    rmse_values = []
    optimal_x_values = []
    cct_sun_values = []
    cct_model_values = []
    eml_sun_values = []
    eml_model_values = []
    synthesized_spds = []

    for i in range(sun_spds.shape[1]):
        b = sun_spds[:, i]
        x, _ = nnls(A, b)
        optimal_x_values.append(x)
        synthesized_spd = np.dot(A, x)
        synthesized_spds.append(synthesized_spd)
        
        # 计算评估指标
        rmse = calculate_rmse(b, synthesized_spd)
        rmse_values.append(rmse)
        
        cct_sun = calculate_cct(wavelengths, b)
        cct_model = calculate_cct(wavelengths, synthesized_spd)
        cct_sun_values.append(cct_sun)
        cct_model_values.append(cct_model)
        
        eml_sun = calculate_eml(wavelengths, b)
        eml_model = calculate_eml(wavelengths, synthesized_spd)
        eml_sun_values.append(eml_sun)
        eml_model_values.append(eml_model)
    
    synthesized_spds = np.array(synthesized_spds).T

    # 3. 结果可视化：RMSE随时间变化的折线图
    plt.figure(figsize=(10, 6))
    plt.plot(sun_times, rmse_values, marker='o')
    plt.title('模型匹配鲁棒性评估：RMSE随时间的变化')
    plt.xlabel('时间')
    plt.ylabel('均方根误差 (RMSE)')
    plt.xticks(rotation=45, ha='right')
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("模型鲁棒性评估图表已生成。")

    # 4. 结果可视化：CCT 和 EML 随时间变化的对比图
    fig, axes = plt.subplots(2, 1, figsize=(10, 10), sharex=True)
    
    axes[0].plot(sun_times, cct_sun_values, marker='o', label='太阳光谱CCT')
    axes[0].plot(sun_times, cct_model_values, marker='x', linestyle='--', label='模型合成光谱CCT')
    axes[0].set_title('相关色温 (CCT) 随时间的变化')
    axes[0].set_ylabel('CCT')
    axes[0].legend()
    axes[0].grid(True, linestyle='--', alpha=0.6)

    axes[1].plot(sun_times, eml_sun_values, marker='o', label='太阳光谱EML')
    axes[1].plot(sun_times, eml_model_values, marker='x', linestyle='--', label='模型合成光谱EML')
    axes[1].set_title('等效褪黑激素照度 (EML) 随时间的变化')
    axes[1].set_xlabel('时间')
    axes[1].set_ylabel('EML')
    axes[1].legend()
    axes[1].grid(True, linestyle='--', alpha=0.6)
    
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
    print("CCT和EML对比图表已生成。")

    # 5. 新增图表：光谱匹配度可视化
    selected_times = ['07:30:00', '12:30:00', '17:30:00']
    time_indices = [sun_times.index(t) for t in selected_times if t in sun_times]
    
    if len(time_indices) == 3:
        fig, axes = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
        fig.suptitle('模型在不同时间的光谱匹配度可视化', fontsize=16)

        for i, idx in enumerate(time_indices):
            ax = axes[i]
            ax.plot(wavelengths, sun_spds[:, idx], label='太阳光谱', color='orange')
            ax.plot(wavelengths, synthesized_spds[:, idx], label='模型合成光谱', color='blue', linestyle='--')
            ax.set_title(f'时间：{sun_times[idx]}')
            ax.set_xlabel('波长 (nm)')
            ax.set_ylabel('光谱功率 (mW/m²/nm)')
            ax.legend()
            ax.grid(True, linestyle='--', alpha=0.6)
        
        plt.tight_layout(rect=[0, 0, 1, 0.96])
        plt.show()
        print("光谱匹配度可视化图表已生成。")
    
    print("-" * 50)

def run_problem4_analysis():
    """
    执行问题4的睡眠指标分析与可视化，并进行统计检验。
    """
    print("--- 正在进行问题4：睡眠指标分析与可视化 ---")
    
    # 1. 读取数据
    file_url = 'https://github.com/Astraeushub/Astraeushub/raw/main/%E9%99%84%E5%BD%95.xlsx'
    try:
        df_raw = pd.read_excel(file_url, sheet_name='Problem 4', header=None)
    except Exception as e:
        print(f"错误：读取文件时出错。请检查URL和工作表名称。错误信息: {e}")
        return

    # 2. 数据预处理 - 转换为长格式，方便分析 (基于您的代码)
    subject_header = df_raw.iloc[0].ffill()
    night_header = df_raw.iloc[1]
    new_columns = [f"{sub}_{night}" for sub, night in zip(subject_header, night_header)]
    df_raw.columns = new_columns
    df_sleep_data = df_raw.iloc[2:].copy()
    df_sleep_data = df_sleep_data.apply(pd.to_numeric, errors='coerce')

    metrics_list = []
    epoch_duration_sec = 30
    condition_mapping = {
        'Night 1': '普通光照',
        'Night 2': '优化光照',
        'Night 3': '黑暗环境'
    }
    
    for col_name in df_sleep_data.columns:
        subject_id, night_str = col_name.split('_')
        condition = condition_mapping.get(night_str, '未知')
        
        sleep_stages = df_sleep_data[col_name].dropna().values
        stages_series = pd.Series(sleep_stages)
        total_epochs = len(stages_series)
        
        # 统计指标
        is_sleep = stages_series.isin([1, 2, 3, 5]) # 假设1,2,3,5为睡眠阶段
        total_bedtime_min = total_epochs * epoch_duration_sec / 60
        
        first_sleep_epoch_idx = stages_series[is_sleep].index[0] if not stages_series[is_sleep].empty else np.nan
        sol = (first_sleep_epoch_idx - stages_series.index[0]) * epoch_duration_sec / 60 if not pd.isna(first_sleep_epoch_idx) else np.nan
        
        tst_epochs = len(stages_series[is_sleep])
        n3_epochs = len(stages_series[stages_series == 3])
        rem_epochs = len(stages_series[stages_series == 5])
        
        tst_min = tst_epochs * epoch_duration_sec / 60
        se = (tst_epochs / total_epochs) * 100 if total_epochs > 0 else 0
        n3_percent = (n3_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        rem_percent = (rem_epochs / tst_epochs) * 100 if tst_epochs > 0 else 0
        
        awakenings = 0
        if not pd.isna(first_sleep_epoch_idx):
            post_sol_stages = stages_series.iloc[int(first_sleep_epoch_idx):]
            awakenings = (post_sol_stages == 4).diff().sum() if (post_sol_stages == 4).any() else 0
            
        metrics_list.append({
            'Subject': subject_id,
            'Condition': condition,
            '总睡眠时间 (分钟)': tst_min,
            '睡眠效率 (%)': se,
            '入睡潜伏期 (分钟)': sol,
            'N3睡眠比例 (%)': n3_percent,
            'REM睡眠比例 (%)': rem_percent,
            '夜间醒来次数': awakenings
        })
    df_metrics = pd.DataFrame(metrics_list)
    
    # 将Condition列转换为有序的分类类型，以确保箱线图的顺序
    order = ['普通光照', '优化光照', '黑暗环境']
    df_metrics['Condition'] = pd.Categorical(df_metrics['Condition'], categories=order, ordered=True)
    df_metrics.sort_values('Condition', inplace=True)

    # 3. 结果可视化：箱线图
    metrics_to_plot = ['总睡眠时间 (分钟)', '睡眠效率 (%)', '入睡潜伏期 (分钟)',
                       'N3睡眠比例 (%)', 'REM睡眠比例 (%)', '夜间醒来次数']
                       
    fig, axes = plt.subplots(2, 3, figsize=(18, 12))
    fig.suptitle('不同光照条件下睡眠指标的分布', fontsize=16)

    for i, metric in enumerate(metrics_to_plot):
        ax = axes.flatten()[i]
        df_metrics.boxplot(column=metric, by='Condition', ax=ax, grid=False, rot=0)
        ax.set_title(metric, fontsize=12)
        ax.set_xlabel('光照条件', fontsize=10)
        ax.set_ylabel(metric, fontsize=10)
        ax.get_figure().suptitle('') # 移除子图自带的suptitle
    
    plt.tight_layout(rect=[0, 0, 1, 0.96])
    plt.show()
    print("睡眠指标箱线图已生成。")

    # 4. 新增图表：睡眠指标对比柱状图
    # 睡眠时间
    plt.figure(figsize=(8, 6))
    avg_tst = df_metrics.groupby('Condition')['总睡眠时间 (分钟)'].mean()
    std_tst = df_metrics.groupby('Condition')['总睡眠时间 (分钟)'].std()
    avg_tst.plot(kind='bar', yerr=std_tst, capsize=4, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
    plt.title('不同光照条件下平均总睡眠时间对比')
    plt.xlabel('光照条件')
    plt.ylabel('平均总睡眠时间 (分钟)')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("平均总睡眠时间柱状图已生成。")

    # 睡眠效率
    plt.figure(figsize=(8, 6))
    avg_se = df_metrics.groupby('Condition')['睡眠效率 (%)'].mean()
    std_se = df_metrics.groupby('Condition')['睡眠效率 (%)'].std()
    avg_se.plot(kind='bar', yerr=std_se, capsize=4, color=['#1f77b4', '#ff7f0e', '#2ca02c'])
    plt.title('不同光照条件下平均睡眠效率对比')
    plt.xlabel('光照条件')
    plt.ylabel('平均睡眠效率 (%)')
    plt.xticks(rotation=0)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.tight_layout()
    plt.show()
    print("平均睡眠效率柱状图已生成。")
    
    # 5. 统计检验（ANOVA & Tukey's HSD）
    print("\n--- 正在进行统计检验 ---")
    for metric in metrics_to_plot:
        print(f"\n--- 对 '{metric}' 进行方差分析 (ANOVA) ---")
        
        group1 = df_metrics[df_metrics['Condition'] == '普通光照'][metric].dropna()
        group2 = df_metrics[df_metrics['Condition'] == '优化光照'][metric].dropna()
        group3 = df_metrics[df_metrics['Condition'] == '黑暗环境'][metric].dropna()
        
        if len(group1) > 1 and len(group2) > 1 and len(group3) > 1:
            f_statistic, p_value = stats.f_oneway(group1, group2, group3)
            print(f"F-statistic: {f_statistic:.4f}")
            print(f"P-value: {p_value:.4f}")
            
            if p_value < 0.05:
                print("结果解读：P-value 小于显著性水平 0.05，结论：三种光照环境对该指标的影响存在显著差异。")
                print("\n--- 进行事后检验 (Tukey's HSD) ---")
                
                data_for_tukey = df_metrics[metric].dropna()
                groups_for_tukey = df_metrics['Condition'].loc[data_for_tukey.index]
                
                tukey_result = pairwise_tukeyhsd(data_for_tukey, groups_for_tukey, alpha=0.05)
                print(tukey_result)
            else:
                print("结果解读：P-value 大于显著性水平 0.05，结论：三种光照环境对该指标的影响不存在显著差异。")
        else:
            print(f"数据不足，无法对 {metric} 进行统计分析。")

    print("-" * 50)

# 主程序入口
if __name__ == '__main__':
    run_problem3_analysis()
    run_problem4_analysis()
```
